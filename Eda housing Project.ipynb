## Function to get the percentage of null values for each column in the dataframe.

```python
def percent_missing(DF):
    percent_null = 100 * df.isnull().sum() / len(df)
    percent_null = percent_null[percent_null > 0].sort_values()
    return percent_null
```

```python
percent_null  = percent_missing(df)
```

```python
percent_null
```

## Below we are visualizing the percentage of null values for each column.

```python
plt.figure(figsize = (8,4),dpi = 200)
sns.barplot(x = percent_null.index,y = percent_null)
plt.xticks(rotation = 90)
```

## Q1- Data Cleaning

```python
# The Numerical Basement columns
```

```python
bsmt_num_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF','Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[bsmt_num_cols] = df[bsmt_num_cols].fillna(0)
```

```python
# The String Basement columns
```

```python
bsmt_str_cols =  ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
df[bsmt_str_cols] = df[bsmt_str_cols].fillna('None')
```

```python
percent_null = percent_missing(df)
```

```python
sns.barplot(x = percent_null.index,y = percent_null)
plt.xticks(rotation = 90)
plt.ylim(0,1)
```

```python
df = df.dropna(axis = 0,subset = ['Electrical','Garage Cars'])
```

```python
percent_null = percent_missing(df)
```

```python
plt.figure(figsize = (8,4),dpi = 200)
sns.barplot(x = percent_null.index,y = percent_null)
plt.xticks(rotation = 90)
plt.ylim(0,1)
```

### Filling the columns with the required values . As we can see that "Mas Vnr Type" has some null values we fill that with 'None' as it was suitable for that feature or column and doing same with the "Mas Vnr Area" we filled thsi column with 0 as it has some null values. 

```python
df["Mas Vnr Type"] = df["Mas Vnr Type"].fillna("None")
df["Mas Vnr Area"] = df["Mas Vnr Area"].fillna(0)
```

```python
percent_null = percent_missing(df)
```

### Below we see that we have decreased the null value columns.
